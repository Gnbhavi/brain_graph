```python
def for_index_finding(n_u, n_v):
    if len(n_u) != len(n_v):
        print("The length of n_u and n_v are not equal")
        return 0, 0, 0, 0, 0
    else:
        product_val = n_u * n_v
        sum_val = n_u + n_v
        sz_index = np.sum(product_val)
        PI_index = np.sum(sum_val)
        ABC_index = np.sum(np.sqrt((sum_val - 2) / product_val))
        NGG_index = np.sum(1/np.sqrt(product_val))
        Mo_index = np.sum(np.abs(n_u - n_v))
        return sz_index, PI_index, ABC_index, NGG_index, Mo_index
```

```python
def file_reader(file_path):
    with open(file_path) as f:
        array_neigh = [x for x in f.read().split('\n')]
        print(len(array_neigh))
    return array_neigh
```

```python
def index_array_creator(outer_file_path):
    n_v_path = os.path.join(outer_file_path, "neigh_v_val.txt")
    n_u_path = os.path.join(outer_file_path, "neigh_u_val.txt")
    n_u_whole_class = file_reader(n_u_path)
    n_v_whole_class = file_reader(n_v_path)
    indices = []
    for i in range(len(n_u_whole_class) - 1):
        try:
            val1 = np.array([int(x) for x in n_u_whole_class[i][1:-1].split(",")])
            val2 = np.array([int(x) for x in n_v_whole_class[i][1:-1].split(",")])        
        except:
            print("Error in line: ", i)  
            break 
        indices.append(for_index_finding(val1, val2))

    return indices
    # return None
```

```python
file_path = "Alzheimers_graph_valued_Dataset 2/test/NonDemented"
```

```python
index_arrays = index_array_creator(file_path)
df = pd.DataFrame(index_arrays, columns=["sz_index", "PI_index", "ABC_index", "NGG_index", "Mo_index"])
```

```python
df.to_csv(os.path.join(file_path, "index_values.csv"))
```

```python
def find_file_in_folders(root_folder, target_file_name):
    index_value = None
    vsl_check = 0
    demented_class = {'NonDemented': 0, 'VeryMildDemented': 1, 'MildDemented': 2, 'ModerateDemented': 3}
    for folder_path, _, files in os.walk(root_folder):
        # print(type(files))
        if target_file_name in files:
            demented_class_key = str(folder_path).split('/')[-1]
            print(str(folder_path).split('/')[-1])
            index_value_creator = pd.read_csv(os.path.join(folder_path, target_file_name))
            index_value_creator = pd.DataFrame(index_value_creator)
            index_value_creator["demented_class"] = demented_class[demented_class_key]
            if vsl_check == 0:
                index_value = index_value_creator
                vsl_check += 1
            else:
                index_value = pd.concat([index_value, index_value_creator], ignore_index=True)
            # return os.path.join(folder_path, target_file_name)
    
    return index_value

root_folder = "Alzheimers_graph_valued_Dataset 2/test"
root_folder2 = "Alzheimers_graph_valued_Dataset 2/train"
target_file_name = "values_of_index_with_weiner.csv"

found_file_path = find_file_in_folders(root_folder, target_file_name)
found_file_path2 = find_file_in_folders(root_folder2, target_file_name)


# if found_file_path:
#     print(f"File '{target_file_name}' found at: {found_file_path}")
# else:
#     print(f"File '{target_file_name}' not found in the specified folder and its subfolders.")
```
