```python
import numpy as np
import Ws_model_generator as wsg
import matplotlib.pyplot as plt
import networkx as nx
```

```python
n=17
k = 4
ws = wsg.Ws_generator(n, k)
graph = ws.G
minimum_distances = dict(nx.shortest_path_length(ws.G))
the_neighbourhood_vertices = {}
```

```python
nx.draw_circular(graph, with_labels=True, node_size=1000, node_color='y')
```

```python
for k_values in range(ws.halfk):
  the_neighbourhood_vertices['n_u_0' + str(k_values+1)] = []
  the_neighbourhood_vertices['n_v_0' + str(k_values+1)] = []
  values_0  = dict(sorted(minimum_distances[0].items()))
  values_k  = dict(sorted(minimum_distances[k_values+1].items()))
  the_neighbourhood_vertices['n_u_0' + str(k_values+1)] = [i for i in range(len(values_0)) if values_0[i] > values_k[i]]
  the_neighbourhood_vertices['n_v_0' + str(k_values+1)] = [i for i in range(len(values_0)) if values_0[i] < values_k[i]]
```

```python
A = int(np.floor((n - 1)/k))
n_v = [0] * (ws.halfk)
mod_rem = np.mod(n - 1, k)
rem_val = [the_rem for the_rem in range(0, int(k/2))]

for i in range(int(k/2)):
    # print(rem_val[:i+1])
    if mod_rem not in rem_val[:i+1]:
        # print("if")
        n_v[i] = (i+1) * A + 1
    else:
        # print("else")
        n_v[i] = (i+1) * A + 1 - (i + 1 - mod_rem)
        # n_v[i] = (i + 1) * A + 1 - (i + 1 - mod_rem + 1)

```
